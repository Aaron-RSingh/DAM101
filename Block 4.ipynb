# Video Lecture 4.4 - Processing Lists and List Comprehension

```python
# TIY Problem 1 - Write a program in Python environment that takes in input of 5 numbers and calculates their average.

# Create an empty list to store the numbers
list1  = []

# Iterate 5 times to get input from the user
for i in range(0, 5):
    new_int = int(input("Enter a number: ")) 
    print("You entered: ", new_int)
    list1.append(new_int)

# Print the list of numbers and calculate the average
print("The numbers you entered are: ", list1)
print("The average is: ", sum(list1) / len(list1))
```

```python
# TIY Problem 2 - Write a program in Python environment that uses list comprehension, which turns [4, 5, 1, 2, 7, 9, 6, 10] into [10, 2, 14, 18]

# Create first list of numbers

list1 = [4, 5, 1, 2, 7, 9, 6, 10]
# Create new list using list comprehension
list2 = [x * 2 for x in list1 if x % 2 != 0]
print("The original list is: ", list1)
print("The new list is: ", list2)
```
