# Video Lecture 6.1 - Sets

```python
# TIY

# Write a program in Python environmentthat takes 10 inputs, stores them in a list and removes any duplicates.

# Hint: use the input() function inside a FOR loop, which repeats 10 times

# Append elements into a list, then convert to a set and convert back

# Create an empty list
list1 = []

# Add 10 numbers to the list from user input
for i in range(10):
    element = int(input("Enter an element: "))
    list1.append(element)

# Convert the list into a set to remove duplicates
unique_elements = set(list1)

# Conver the set back into a unique list
unique_list = list(unique_elements)

# Print the unique list
print("Unique elements:", unique_list)
```

# Video Lecture 6.2 - Built in Functions for Sets and Set Methods

```python
# TIY

# Write a program that keeps taking characters as the input and storign them in a set until empty string is entered.

# Then print the number of unique characters that was entered.

# Note: use the input() function inside a WHILE loop. User the BREAK statement to exit the loop when "" is encountered.

# Create an empty set to store unique characters
unique_characters = set()

while True:
    char = input("Enter a character (or press Enter to stop): ")
    if char == "":
        break
    unique_characters.add(char)

print("Number of unique characters:", len(unique_characters))
print("Unique characters entered:", unique_characters)
```

# Video Lecture 6.3 - Set Operations and Processing Sets

```python
# TIY

# Using the two provided sets, write a program which determines the following:

# Items bought by both customers1 and customer2
# All items bought by customer1 and customer2
# Items that customer1 or customer 2 but not bought by both
# Items that customer2 bough but not customer1
# Remaining items the customer1 had after they returned the items matching with customer2

customer1 = {"oil paints", "lamp", "watch", "frame", "kit-kat", "desk"}
customer2 = {"lamp", "monitor", "home-hub", "desk", "canvas"}

# Items bought by both customers1 and customer2
both_customers = customer1.intersection(customer2)
print("Items bought by both customers:", both_customers)

# All items bought by customer1 and customer2
all_items = customer1.union(customer2)
print("All items bought by customer1 and customer2:", all_items)

# Items that customer1 or customer 2 but not bought by both
unique_items = customer1.symmetric_difference(customer2)
print("Items that customer1 or customer2 but not bought by both:", unique_items)

# Items that customer2 bought but not customer1
customer2_unique = customer2.difference(customer1)
print("Items that customer2 bought but not customer1:", customer2_unique)

# Remaining items the customer1 had after they returned the items matching with customer2
customer1_remaining = customer1.difference(customer2)
print("Remaining items customer1 had after returning items matching with customer2:", customer1_remaining)
```

```python
# TIY 2

# Problem 2: Write a program which uses a dictionary to count the frequency of numbers in a list.

numbers = []

x = int(input("How many numbers do you want to enter? "))

for i in range(x):
    num = int(input("Enter a number: "))
    if num == "":
        break
    numbers.append(num)

frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(sorted(frequency.items()))
```

```python
# Project: School Musical Sign-Ups

# Create a program that gives those interested in auditioning the ability to sign up for their preferred role. Only 5 students can sign up for each day of auditions decided on a first come first served basis.

# Roles: Principal, Teacher, Troublemaker, Students

# Program prompts: 1. What is your name? 2. What is your grade? 3. What is your preferred role?

# Create a dictionary to store the response

from os import name


auditions = {
    "Principal": {
        },
    "Teacher": {
        },
    "Troublemaker": {
        },
    "Students": {
        }
}

def sign_up():
    name = input("What is your name?").capitalize()
    grade = int(input("What is your grade? Please enter a number"))
    role = int(input("""Please select the number of your preferred role from the following: 
                    1. Principal
                    2. Teacher 
                    3. Troublemaker
                    4. Students
                    """))
    if role == '1':
        auditions['Principal'][name] = grade
    elif role == '2':
        auditions['Teacher'][name] = grade
    elif role == '3':
        auditions['Troublemaker'][name] = grade
    else:
        auditions['Students'][name] = grade
# For-loop that calls the sign_up() function 5 times
for i in range(5):
    sign_up()
# Printout for the list of students signed up to audition
print("Sign-ups for 'A Day without a Principal' are now closed")
print("Role: Principal")
for student, grade in auditions['Principal'].items():
    print(student, grade)
print("Role: Teacher")
for student, grade in auditions['Teacher'].items():
    print(student, grade)
print("Role: Troublemaker")
for student, grade in auditions['Troublemaker'].items():
    print(student, grade)
print("Role: Students")
for student, grade in auditions['Students'].items():
    print(student, grade)

print(auditions)
```

# Video Lecture 6.8 - Write Data to a File

```python
# TIY

# Write a program in pyhton environment that 1. reads in a file, 2. Calculates the number of characters, words and lines in the file, 3. Writes this information to a different text file.

# Open the input file for reading
input_file = open("input6_8.txt", "r")
# Read the contents of the input file
contents = input_file.readlines()
# Close the input file
input_file.close()

# Calculate the number of characters, words and lines
num_characters = sum(len(line) for line in contents)
num_words = sum(len(line.split()) for line in contents)
num_lines = len(contents)
# Open the output file for writing
output_file = open("output6_8.txt", "w")
# Write the information to the output file
output_file.write(f"Number of characters: {num_characters}\n")
output_file.write(f"Number of words: {num_words}\n")
output_file.write(f"Number of lines: {num_lines}\n")
# Close the output file
output_file.close()
```
