```python
# Problem: Calculate the area and circumference of a circle with user input radius

# Forumula for area of a circle: A = πr^2
# Forumula for circumference of a circle: C = 2πr
# Note: π is a constant that is approximately equal to 3.14

pi = 3.14

def calculate_circle_area(radius):
    area = pi * radius ** 2
    return area

def calculate_circle_circumference(radius):
    circumference = 2 * pi * radius
    return circumference

radius = float(input("Enter the radius of the circle: "))

#area = calculate_circle_area(radius)
#circumference = calculate_circle_circumference(radius)
print(radius)
print("The area of the circle is: {:.2f}".format(calculate_circle_area(radius)))
print("The circumference of the circle is: {:.2f}".format(calculate_circle_circumference(radius)))
```

```python
# TIY 1

# Write a function in python environment that determimes how many days there are in a particular month and the input should be read as the month name (assuming a non-leap year) in the main program

def days_in_month(month, year):
    month = month.lower()  # Convert the input to lowercase for case-insensitive comparison
    if month in ["january", "march", "may", "july", "august", "october", "december"]:
        return 31
    elif month in ["april", "june", "september", "november"]:
        return 30
    elif month == "february":
        if (year % 4 == 0):
            return 29
        return 28
    else:
        return None  # Return None for invalid month names
    
month_name = input("Enter the name of the month: ")
year = int(input("Enter the year: "))
days = days_in_month(month_name, year)
if days is not None:
    print("There are {} days in {}.".format(days, month_name))
else:
    print("Invalid month name entered.")

# TIY 2

# Extend the previous exercise by including a second parameter in the function passing in the year.

# Then, determine whether the year is a leap year or not and subsequenetly adjust the value of the number of days in February accordingly


```

```python
# Project: Customer Service Bot

# Tiny Space is a company that needs to create a chatbot to speak with customers. Design the chat bot.

# The chatbot should be able to:
# - Greet the customer
# Inquiry Catergory and assign to the correct department
# Note: Store location and hours is the only option that doesn't get sent to a team.
# The user should input a number from 1-5 depending on the category of their inquiry:
# - Store Location and Hours
# - Order Status: Elliot
# - Issue with Order: Chrissa
# - Design Services: Ramon
# - Other: Trinity

def more_help():
    print("Is there anything else I can assist you with? (yes/no)")
    response = input().lower()
    if response == "yes":
        chatbot()
    else:
        print("Thank you for contacting Tiny Space Customer Service. Have a great day!")

def chatbot():
    print("Welcome to Tiny Space Customer Service!")
    print("Please select the category of your inquiry:")
    print("""
        1. Store Location and Hours
        2. Order Status
        3. Issue with Order
        4. Design Services
        5. Other
          """)

    category = input("Enter the number corresponding to your inquiry: ")


    if category == "1":
        print("Our store is located at 2300 Riverdale Lane, Boston, MA 02101 and is open Monday to Saturday from 10 a.m. to 6 p.m.")
        more_help()
    elif category == "2":
        print("Sure, I can help you with that.")
        full_name = input("Please enter the full name on your order: ")
        order_number = input("Enter your order number: ")
        print(f"""
              Name: {full_name}
              Order Number: {order_number}
              """)
        print("Awesome! I'm checking on the status of the order now.")
    elif category == "3":
        print("I'm sorry that you're experiencing issues with your order.")
        full_name = input("Please enter the full name on the order: ")
        order_number = input("Enter your order number: ")
        issue_description = input("Please describe the issue you're having with your order: ")
        print(f"""
              Name: {full_name}
              Order Number: {order_number}
              Issue: {issue_description}
              """)
        print("Thanks for providing that information. I'm looking into this now.")
    elif category == "4":
        print("I can definitely help you out with your design questions! Tell me how I may be of assistance.")
        design_question = input("Please enter your design question: ")
        print(f"""
              Design Question: {design_question}
              """)
        print("Thanks for sharing your design question. I'm looking into this now.")
    elif category == "5":
        print("No problem, please describe to me how I may be of assistance.")
        other_inquiry = input("Please enter your inquiry: ")
        print(f"""
              Inquiry: {other_inquiry}
              """)
        print("Thanks for sharing your inquiry. I'm looking into this now.")
    else:
        print("Invalid input. Please enter a number from 1 to 5.")
        chatbot()  # Restart the chatbot for valid input

chatbot()
```
