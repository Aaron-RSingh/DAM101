```python
# Problem: Calculate the area and circumference of a circle with user input radius

# Forumula for area of a circle: A = πr^2
# Forumula for circumference of a circle: C = 2πr
# Note: π is a constant that is approximately equal to 3.14

pi = 3.14

def calculate_circle_area(radius):
    area = pi * radius ** 2
    return area

def calculate_circle_circumference(radius):
    circumference = 2 * pi * radius
    return circumference

radius = float(input("Enter the radius of the circle: "))

#area = calculate_circle_area(radius)
#circumference = calculate_circle_circumference(radius)
print(radius)
print("The area of the circle is: {:.2f}".format(calculate_circle_area(radius)))
print("The circumference of the circle is: {:.2f}".format(calculate_circle_circumference(radius)))
```

```python
# TIY

# Write a program in python environment that:

# 1. Reads in a string of numbers separated by spaces and stores the numbers in a list
# 2. List is then passed to a function distinct_numbers to create a list of distinct numbers
# 3. The list is then returned to the main program for printing
# 4. For example: if input = 1 2 4 5 2 3 1 2 3 6, then output = 1 2 4 5 3 6


distinct_list = []

def distinct_numbers(numbers):
    for number in numbers:
        if number not in distinct_list:
            distinct_list.append(number)
    return distinct_list

numbers_input = input("Enter a string of numbers separated by spaces: ")
numbers_list = numbers_input.split()
result = distinct_numbers(numbers_list)
print("Distinct numbers:", " ".join(result))

```

```python
# TIY

# Write a program in python environment that takes a string and a character as an input and finds whether the character is found within the string using linear search.

# E.g. String = "Hello World", Character = "o" => Output: "The character 'o' is found in the string 'Hello World'."

def linear_search(string, character):
    for char in string:
        if char == character:
            return True
    return False

input_string = input("Enter a string: ")
input_character = input("Enter a character to search for: ")

if linear_search(input_string, input_character):
    print(f"The character '{input_character}' is found in the string '{input_string}'.")
else:
    print(f"The character '{input_character}' is not found in the string '{input_string}'.")

```

```python
# TIY

# Write a program in python environment that takes an alphabetically sorted string and a character as an input and finds whether the character is found within the string using an ordered linear search.

input_string = input("Enter an alphabetically sorted string: ")

input_character = input("Enter a character to search for: ")


def ordered_linear_search(string, character):
    for char in string:
        if char == character:
            return True
        elif char > character:  # Since the string is sorted, we can stop searching if we pass the character
            break
    return False


if ordered_linear_search(input_string, input_character):
    print(f"The character '{input_character}' is found in the string '{input_string}'.")
else:
    print(f"The character '{input_character}' is not found in the string '{input_string}'.")

```

```python
# Set the index for the search to default value
index = -1

# Ask the user for input
word = input("Please enter a word: ")
character = input("Please enter a character: ")

# Define a sorted list to be searched
list1 = sorted(word)

# Go through each element in word
for i in range(len(list1)):
    # Check if the element is equal to character
    if list1[i] == character:
        # Set the index variable to this element's index and stop the search
        index = i
        break
    # Check if the element is greater than character
    elif list1[i] > character:
        # Stop the search
        break

# Print the answer to the console
if index != -1:
    print("The character {} is found in the string {}.".format(character, word))
else:
    print("The character {} is not found in the string {}.".format(character, word))
```

```python
# Write a program in python environment that takes an alphabetically sorted string and a character as an input and finds whether the character is found within the string using binary search.

def binary_search(string, character):
    first = 0
    last = len(string) - 1

    while first <= last:
        index = (first + last) // 2
        if string[index] == character:
            return True
        elif string[index] > character:
            last = index - 1
        else:
            first = index + 1

    return False

input_string = input("Enter an alphabetically sorted string: ")
input_character = input("Enter a character to search for: ")

if binary_search(input_string, input_character):
    print(f"The character '{input_character}' is found in the string '{input_string}'.")
else:
    print(f"The character '{input_character}' is not found in the string '{input_string}'.")

```

```python
# Get user input
input_string = input("Enter an alphabetically sorted string: ")
input_character = input("Enter a character to search for: ")

# Initialize variables for the search
first = 0
last = len(input_string) - 1
found = False

# The Binary Search Logic
while first <= last:
    # Find the middle point
    index = (first + last) // 2
    
    # Check if the character is at the middle
    if input_string[index] == input_character:
        found = True
        break  # Exit the loop immediately if found
    
    # If character is smaller, ignore the right half
    elif input_string[index] > input_character:
        last = index - 1
        
    # If character is larger, ignore the left half
    else:
        first = index + 1

# Final Output
if found:
    print(f"The character '{input_character}' is found in the string '{input_string}'.")
else:
    print(f"The character '{input_character}' is not found in the string '{input_string}'.")
```
