4) Your manager gives you the following task:

| Starter                       | Main                  | Desert            |
| -----------                   | -----------           | -----------       |
| Calamari rings (6.25)         | Grilled Salmon (12.60)| Ice Cream (3.60)  |
| Chicken Wings (6.50)          | Beef Burger (11.50)   | Lemon Tart (4.75) |
| Grilled Halloumi (5.50)       | Vegetable Lasagne (51)| Cheesecake (5.25) |

i) Write a program that has three dictionaries as shown in a table above. In each dictionary, the key should be the item's name and the value should be the price.

ii) Now create a list called 'menu'. The list should have the two dictionaries from part i) as its elements.

iii) Write a function that takes a list of dictionaries as an argument and ask the user to input a string. The function should then find and return the price of that dish in the menu. The function should use a nested for loop to do the processing. If the dish name is not found in the menu, then the appropriate message should be displayed in the console and 'None' should be returned by the function.

iv) Write code to run the function written in part iii) using the 'menu' list created in part ii).

v) For parts i), ii), iii) and iv) write efficient code and document it as detailed in the Documentation Requirements.

```python
# i) & v)

# Write a program that has three dictionaries as shown in a table above. In each dictionary, the key should be the item's name and the value should be the price.


# Created three separate dictionaries for starters, mains and desserts with the food name as the key and the price given as the value.
starter_dictionary = {"Calamari rings": 6.25, "Chicken Wings": 6.50, "Grilled Halloumi": 5.50}
main_dictionary = {"Grilled Salmon": 12.60, "Beef Burger": 11.50, "Vegetable Lasagne": 51}
dessert_dictionary = {"Ice Cream": 3.60, "Lemon Tart": 4.75, "Cheescake": 5.25}

# Test to see if the three dictionaries have been created correctly by printing them out
print(starter_dictionary, main_dictionary, dessert_dictionary)
```

```python
# ii) & v)

# Now create a list called 'menu'. The list should have the two dictionaries from part i) as its elements.


# Created a list called 'menu' and added the three previously created dictionaries as elements in the list. The question states 2 dictionaries but from the responses in the QandA, it seems this was a mistake and that all three dictionaries should be added to the list.
menu = [starter_dictionary, main_dictionary, dessert_dictionary]

# Test to see if the list has been created correctly by printing it out
print(menu)
```

```python
# iii) & v)

# Write a function that takes a list of dictionaries as an argument and ask the user to input a string. The function should then find and return the price of that dish in the menu. The function should use a nested for loop to do the processing. If the dish name is not found in the menu, then the appropriate message should be displayed in the console and 'None' should be returned by the function.

def find_price(menu):
    """
    This function takes a users input of a dish name and checks against a list of dictionaries to find the price of the dish.

    The one required argument is the menu, which is a list of 3 dictionaries that contain 3 dishes and their corresponding prices for each dictionary.

    The return value is the price of the dish if it is found in the menu, otherwise it returns None and prints a message saying "Dish not found in the menu."     
    """
    
    # Request user input for the name of dish to look up the price
    dish_name = input("Please enter the name of the dish: ")
    
    # Nested for loop to cross check the user input dish against the 3 dictionaries in the menu list 
    for category in menu:
        for dish, price in category.items():

            # Check if the dish name matches the user input, ignoring case sensitivity. If successful match, the price is returned
            if dish.lower() == dish_name.lower():
                return price
    
    # If the dish name is not found in any of the dictionaries, "Dish not found in the menu" message is printed and None is returned
    print("Dish not found in the menu.")
    return None

# Test to see if the functions works by calling the function with the menu list as the argument
find_price(menu)
```

```python
# iv) & v)

# Write code to run the function written in part iii) using the 'menu' list created in part ii).

# Run the most recent find_price function of the modified function using the menu list as the argument
find_price(menu)
```
