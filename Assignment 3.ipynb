```python
# ii) & iv)

# Extend the function in part i) to also use the insertion sort algorithm to sort the list in descending order.

def read_file(filename):

    """
    This function reads a text file and returns a list of integers separated by spaces or newlines in descending order using the insertion sort algorithm

    The one argument required is a string that is the name of the text file to be read. The text file should be in the same location as the Python file that contains this function. The text file should contain integers separated by spaces and/or newlines. This function also handles potential blank lines in the text file to prevent errors

    The return value is a list of integers in descending orderthat were read from the provided text file 
    """

    # Open the file and read the file line by line to be stored in the variable 'contents'
    file = open(filename, 'r')
    contents = file.read()

    # Close the data file after reading
    file.close()
    
    # Split the contents of the file into a list of strings
    single_list = []
    for item in contents.split():
        
        # Check if the item is not empty (I wasn't sure if the test case had a typo or a blank line so I added this check to prevent potential errors)
        if item:

            # Convert the string to an integer and append it to the list
            single_list.append(int(item))
    
    # Sort the list in descending order using the insertion sort algorithm
    for i in range(1, len(single_list)):
        key = single_list[i]
        j = i - 1
        while j >= 0 and single_list[j] < key:
            single_list[j + 1] = single_list[j]
            j -= 1
        single_list[j + 1] = key

    # Return the descending list of integers
    return single_list

# Print the output of the modified function using the input.txt test case provided
print(read_file('input.txt'))
```
