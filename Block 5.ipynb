# Video Lecture 5.4 - String Functions and Methods

```python
# TIY

# Create an empty list to append countries into
country_list = []

# Set a counter to keep track of how many countries need to be added
count = 0

# Use a while loop to prompt user to enter a country until 3 countries have been added
while count < 3:
# Convert country into string and force it to be lowercase before appending to country list
    country = str(input("Enter a country: ")).lower()
    country_list.append(country)
# Increase count incrementally until 3 countries have been added
    count += 1

# Sort the list of countries in alphabetical order
country_list.sort()
for i in range(len(country_list)):
# Capitalise the first letter of each country in the list
    country_list[i] = country_list[i].capitalize()

# Print the output of the countries list 
print("The countries you entered are:", country_list)
```

# Video Lecture 5.5 - Substrings and String Formatting

```python
# Create a list of vowels
vowels = ['a', 'e', 'i', 'o', 'u']

# Get a lowercase letter from user input
letter = input("Please enter a letter: ").lower()

# Check if the input was a not single letter 
# and it was not, print invalid input
if not (letter.isalpha() and len(letter) == 1):
    print("Invalid input")
else:
    # Check if letter is a vowel
    print("Letter '{}' is a {}".format(
        letter, "vowel" if letter in vowels else "consonant"))
```

# Video Lecture 5.8 - Tuple Processing

```python
# TIY

# Problem 1: Write a program which stores a list of 7 numbers in a tuple.
# Then determines the average car price.

#Hint: Use an input() funtion within a loop that executes 7 times. Then conhvert the list into a tuple.

# Calculate the average by using sum() and len() functions.

car_prices = []
for i in range(7):
    price = float(input("Enter a car price: "))
    car_prices.append(price)

print(car_prices)

car_prices_tuple = tuple(car_prices)

print(car_prices_tuple)
print(sum(car_prices_tuple))
print(len(car_prices_tuple))

average_price = sum(car_prices_tuple) / len(car_prices_tuple)
print(f"The average car price is: {average_price}")
```
