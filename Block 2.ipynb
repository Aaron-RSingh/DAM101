```python
# Video Lecture 2.8 - Nester IF Statements

# determine what day of the week it is first
day_of_week = input("what day of the week is it? ").lower()

# set variable for output
x = ""

# initially asked both questions about weekday and time of day however not relevant if weekend is chosen
# time_of_day = int(input("what hour of the day is it? Please use integers between 0-24. "))

# check if weekday or weekend and provide output if weekend is chosen
if (day_of_week == "saturday" or day_of_week == "sunday") :
    x = "The bank branch is closed."
elif (day_of_week == "monday" or day_of_week == "tuesday" or day_of_week == "wednesday" or day_of_week == "thursday" or day_of_week == "friday") :

# once confirmed weekday, check what the hour the day is in using a 24 "hour" checking
    time_of_day = int(input("what hour of the day is it? Please use integers between 0-24. "))
    if (time_of_day >= 9 and time_of_day < 17) :

# update output variable depending on what time of day it is
        x = "The bank branch is open."
    else :
        x = "The bank branch is closed."
else :
    x = "That is not a day of the week"

# print output variable
print(x)
```

```python
# Block 2 - Input, Output and IF Statments - Reading 2.8 - Kortext

# Project: What to Wear

# Description: Madison wants to create a program that tells her what she shoudl wear based on the weather conditions.

# 1. If the temperature is 80 degrees or warmer, then Madison should wear shorts and pack her sunglasses.

# 2. If the temperature is 60-79 degrees, then Madison should wear a light jacket.

# 3. If the temperature is 59 degrees of cooler, then Madison should wear a coat in addition to a hat, gloves, and scarf.

# Create conditional logic using if-elif-else statements to help Madison create her program.



# create input variable for the temperature
temperature = int(input("what is the temperature today? "))

# set output variable
outfit = ""

# create conditional if statement
if (temperature >= 80) :
  outfit = "shorts and pack sunglasses."
elif (temperature >= 60 and temperature <= 79) :
  outfit = "a light jacket."
else :
  outfit = "a coat, hat, gloves, and scarf."

variable_output = (f'Madison should wear {outfit}')

# print variable
print(variable_output)
```
