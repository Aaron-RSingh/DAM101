```python
# i) & v)

# Create the list called 'menu' like in Q4 parts i) and ii). Then modify it to be a dictionary instead of a list. Write a function to print the contents of the 'menu' dictionary to the console in an organised and easy to read format.

# Created a dictionary called menu with the three previous categories, which the course type as the key and the respective dishes and prices as a second key/value pair in a nested dictionary
menu = {
    "starters": {
        "Calamari rings": 6.25,
        "Chicken Wings": 6.50,
        "Grilled Halloumi": 5.50
        },
    "mains": {
        "Grilled Salmon": 12.60,
        "Beef Burger": 11.50,
        "Vegetable Lasagne": 51
        },
    "desserts": {
        "Ice Cream": 3.60,
        "Lemon Tart": 4.75,
        "Cheescake": 5.25
        }
}

# Created a function to print the contents of the menu dictionary in an organised and easy to read format. The function uses a for loop to iterate through the categories and dishes and prints them out
def print_menu(menu_dict):
    """
    This function prints out the contents of a dictionary in an organised and easy to read format. The dictionary should have a course type as the key and the respective dishes and prices as a second key/value pair.

    This function takes 1 argument, which is the menu dictionary.

    There is no return value for this function, it simply prints the contents of the menu dictionary to the console.
    """
    
    # Iterate through the menu dictionary and print out the categories and dishes
    for category, dishes in menu_dict.items():
        
        # Print the category name
        print(f"\n{category.capitalize()}:")
        
        # Iterate through the dishes and print out the dish name and price
        for dish, price in dishes.items():
            print(f"  {dish}: {price:.2f}")

# Run the function by printing the menu
print_menu(menu)
```

```python
# ii)

# Write a function called 'add_dish' that has three parameters: 'course', 'name' and 'price'. The function should then add the new entry to a dictionary in 'menu' specified by 'course' parameter. The function should let the user know that the item has been added and print the updated 'items' to the console. If the 'course' parameter is invalid, then an appropriate message should be printed to the console.

def add_dish(course, name, price):
    """
    This function adds a new dish to the menu dictionary under the specified course.

    The function takes three parameters: 'course', which specifies the course type (e.g. 'starters', 'mains', 'desserts'), 'name', which is the name of the dish, and 'price', which is the price of the dish.

    The function does not return any value, it simply updates the menu dictionary and prints the updated menu to the console. If the course parameter is invalid, it prints an appropriate message to the console.
    """

    # Check if the course parameter is valid
    if course in menu:
        
        # Check if the name parameter is already in the menu dictionary for the specified course
        if name in menu[course]:
            print(f"{name} is already on the {course} menu with a price of {menu[course][name]:.2f}.")
            return
        
        # Add the new dish to the menu dictionary
        menu[course][name] = price
        
        # Let the user know that the item has been added and print the updated items to the console
        print(f"{name} has been added to the {course} menu with a price of {price:.2f}.")
        print_menu(menu)
    
    else:
        
        # If the course parameter is invalid, print an appropriate message to the console
        print(f"Invalid course: {course}. Please choose from 'starters', 'mains' or 'desserts'.")

# Test the function by adding a new dish to the menu
add_dish('mains', 'Grilled Salmon', 12.60)
```
