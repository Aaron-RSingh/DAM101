```python
# i) & v)

# Create the list called 'menu' like in Q4 parts i) and ii). Then modify it to be a dictionary instead of a list. Write a function to print the contents of the 'menu' dictionary to the console in an organised and easy to read format.

# Created a dictionary called menu with the three previous categories, which the course type as the key and the respective dishes and prices as a second key/value pair in a nested dictionary
menu = {
    "starters": {
        "Calamari rings": 6.25, # I intentionally left rings in lower case as this was how it was written in the exam questions, as I am not sure whether you might check with the provided names or not, I have added case sensitive checks to each subsequent function to ensure that the menu can be read and updated regardless of the case used by the user when searching, adding or calculating.
        "Chicken Wings": 6.50,
        "Grilled Halloumi": 5.50
        },
    "mains": {
        "Grilled Salmon": 12.60,
        "Beef Burger": 11.50,
        "Vegetable Lasagne": 51
        },
    "desserts": {
        "Ice Cream": 3.60,
        "Lemon Tart": 4.75,
        "Cheescake": 5.25
        }
}

# Created a function to print the contents of the menu dictionary in an organised and easy to read format. The function uses a for loop to iterate through the categories and dishes and prints them out
def print_menu(menu_dict):
    """
    This function prints out the contents of a dictionary in an organised and easy to read format. The dictionary should have a course type as the key and the respective dishes and prices as a second key/value pair.

    This function takes 1 argument, which is the menu dictionary.

    There is no return value for this function, it simply prints the contents of the menu dictionary to the console.
    """
    
    # Iterate through the menu dictionary and print out the categories and dishes
    for category, dishes in menu_dict.items():
        
        # Print the category name
        print(f"\n{category.capitalize()}:")
        
        # Iterate through the dishes and print out the dish name and price. The price is formatted to 2 decimal places. The dish name is printed in title case and due to any possible confusion regarding the Calamari rings or any future case sensitive issues. 
        for dish, price in dishes.items():
            print(f"  {dish.title()}: {price:.2f}")

# Run the function by printing the menu
print_menu(menu)
```

```python
# ii) & v)

# Write a function called 'add_dish' that has three parameters: 'course', 'name' and 'price'. The function should then add the new entry to a dictionary in 'menu' specified by 'course' parameter. The function should let the user know that the item has been added and print the updated 'items' to the console. If the 'course' parameter is invalid, then an appropriate message should be printed to the console.


# Uncomment the following menu dictionary if you want to test this function independently of the previous cells, otherwise make sure to execute the previous cells to have the menu dictionary and print_menu function available for this function to work.

# menu = {
#     "starters": {
#         "Calamari rings": 6.25,
#         "Chicken Wings": 6.50,
#         "Grilled Halloumi": 5.50
#         },
#     "mains": {
#         "Grilled Salmon": 12.60,
#         "Beef Burger": 11.50,
#         "Vegetable Lasagne": 51
#         },
#     "desserts": {
#         "Ice Cream": 3.60,
#         "Lemon Tart": 4.75,
#         "Cheescake": 5.25
#         }
# }


def add_dish(course, name, price):
    """
    This function adds a new dish to the menu dictionary under the specified course.

    The function takes three parameters: 'course', which specifies the course type (e.g. 'starters', 'mains', 'desserts'), 'name', which is the name of the dish, and 'price', which is the price of the dish.

    The function does not return any value, it simply updates the menu dictionary and prints the updated menu to the console. If the course parameter is invalid, it prints an appropriate message to the console.
    """
    
    # Convert the course parameter to lower case to accomodate both possibilities, as mentioned in the comment above
    course = course.lower() 

    # Check if the course parameter is valid
    if course in menu:
        
        # Set a variable to check if the food item already exists in the menu
        exists = False
        
        # Set a variable to store the matched key if the item already exists
        matched_key = None

        # Iterate through items in the specified course to check if item already exists
        for item in menu[course]:
            
            # Check if the name parameter is already in the menu dictionary for the specified course, convert both user input and menu keys to lower case for the comparison to accomodate both possibilities, as mentioned in the comment above
            if item.lower() == name.lower():
                exists = True
                matched_key = item
                # If the item already exists, print the current price of the item to the console in addition to the rest of the menu
                print(f"{name} is already on the {course} menu with a price of {menu[course][matched_key]:.2f}.")
                print_menu(menu)
                return

            # If the item does not exists, add the new dish to the menu dictionary and print the updated menu
        if not exists:
            
            # Conver the name to title case to ensure that the menu is consistent regardless of how the user inputs the name parameter
            name = name.title()

            # Add the new dish to the menu dictionary
            menu[course][name] = price

            # Let the user know that the item has been added and print the updated items to the console
            print(f"{name} has been added to the {course} menu with a price of {price:.2f}.")
            print_menu(menu)
    
    else:
        
        # If the course parameter is invalid, print an appropriate message to the console
        print(f"Invalid course: {course}. Please choose from 'starters', 'mains' or 'desserts'.")
        print_menu(menu)

# Test the function by adding a new dish to the menu
add_dish('Mains', 'beef burger', 9.60)
```

```python
# iii) & v)

# Write a new function that uses the function from Q4 part iii) to ask the user to enter three values in the console: one choice for each course. The function should then calculate and return the total amount for the chosen meal and print the appropriate message to the console. If the user enters a missing name or chooses multiple of the same course, then the message should be printed to the console letting the user know that an invalid selection was made and return 'None'.



# Uncomment the following menu dictionary if you want to test this function independently of the previous cells, otherwise make sure to execute the previous cells to have the menu dictionary and print_menu function available for this function to work.

# menu = {
#     "starters": {
#         "Calamari rings": 6.25,
#         "Chicken Wings": 6.50,
#         "Grilled Halloumi": 5.50
#         },
#     "mains": {
#         "Grilled Salmon": 12.60,
#         "Beef Burger": 11.50,
#         "Vegetable Lasagne": 51
#         },
#     "desserts": {
#         "Ice Cream": 3.60,
#         "Lemon Tart": 4.75,
#         "Cheescake": 5.25
#         }
# }


def calculate_total():
    """
    This function calculates the total amount for a chosen meal based on user input.

    The function takes no arguments, but will request the user to enter their choice of dishes for each course (3 separate inputs). This does require the previous cells of functions to be executed to have the menu dictionary and the print menu function available. Additionally, this function will also work on the updated menu dictionary after a user has added a new dish using the add_dish function.

    The function will return the total price of the meal based on the user's choices. If the user tpyes a dish that is not on the menu, the function will return a message stating an invalid selection was made and return 'None'.
    """

    # Initialize an empty dictionary to store the user's choices
    user_choices = {}

    # Prompt the user to enter their choice for each course
    for course in menu.keys():
        choice = input(f"Please enter your choice for {course}: ")
        user_choices[course] = choice

    # Check if any of the choices are missing or if there are multiple of the same course
    if any(choice == "" for choice in user_choices.values()):
        print("Invalid selection: Missing name. Please make sure to enter a choice for each course.")
        return None
    
    if len(set(user_choices.values())) < len(user_choices):
        print("Invalid selection: Multiple of the same course. Please make sure to choose different dishes for each course.")
        return None

    # Calculate the total price of the meal
    total_price = 0
    for course, choice in user_choices.items():
        
        # Set variable for matched key
        matched_dish = None
        for dish in menu[course]:
            if dish.lower() == choice.lower():
                matched_dish = dish
                break

        # Check if a match was found during the loop
        if matched_dish:
            total_price += menu[course][matched_dish]
        
        # If no match was found, print invalid selection message
        else:
            print(f"Invalid selection: {choice} is not on the {course} menu.")
            return None

    # Print the total amount for the chosen meal stating what items were chosen for each course. I've used the user choices dictionary using join and title,  instead of creating a new variable to store the matched dishes as I am not sure which would be considered more correct
    print(f"The total amount for your meal for {', '.join(user_choices.values()).title()} is: {total_price:.2f}")

# Test the function by executing it to prompt user answers
calculate_total()
```
