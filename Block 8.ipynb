# Video Lecture 8.1 - Introduction to Sorting Algorithms

```python
# This function sorts the list using a simple sorting algorithm
def simple_sort(list1):    
    n = len(list1)
    list_sorted = []
    for i in range(n):
        min1 = list1[0]
        for j in range(len(list1)):
            if list1[j] < min1:
                min1 = list1[j]
        list_sorted.append(min1)
        list1.remove(min1)
    list1 = list_sorted
    return list1

# Ask user for input
input_string = input("Please enter your string: ")

#Sort the string in alphabetical order using the simple_sort function
sorted_list = simple_sort(list(input_string))

# Convert from a list to a string
sorted_string = ""
for i in sorted_list:
    sorted_string += i

#Print the answer
print(sorted_string)
```

# Video Lecture 8.2 - Bubble Sort

```python
# This function sorts the list using a bubble sorting algorithm
def bubble_sort(list1):    
    n = len(list1)
    for i in range(n-1):
        for j in range(n-1-i):
            if list1[j] > list1[j+1]:
                temp = list1[j]
                list1[j] = list1[j+1]
                list1[j+1] = temp
    return list1

# Ask user for input
input_string = input("Please enter your string: ")

#Sort the string in alphabetical order using the simple_sort function
sorted_list = bubble_sort(list(input_string))

# Convert from a list to a string
sorted_string = ""
for i in sorted_list:
    sorted_string += i

#Print the answer
print(sorted_string)
```

# Video Lecture 8.3 - Merge Sort

```python
# This function sorts the list using a merge sort algorithm 
def merge_sort(list1):    
    if len(list1) > 1:
        mid = len(list1) // 2   # List's mid point
        left = list1[:mid]      # Left half of the list
        right = list1[mid:]     # Right half of the list
        merge_sort(left)        # Apply merge_sort to the left sub-list 
        merge_sort(right)       # Apply merge_sort to the right sub-list
        
        # Merge the elements from left and right into correct positions
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                list1[k] = left[i]
                i += 1
            else:
                list1[k] = right[j]
                j += 1
            k += 1
        
        # If left or right has no more elements, 
        # insert the rest into the list
        while i < len(left):
            list1[k] = left[i]
            i += 1
            k += 1
        
        while j < len(right):
            list1[k] = right[j]
            j += 1
            k += 1
            
    return list1

# Ask user for input
input_string = input("Please enter your string: ")

#Sort the string in alphabetical order using the simple_sort function
sorted_list = merge_sort(list(input_string))

# Convert from a list to a string
sorted_string = ""
for i in sorted_list:
    sorted_string += i

#Print the answer
print(sorted_string)
```

# Video Lecture 8.4 - Insertion Sort

```python
# This function sorts the list using a insertion sort algorithm
def insertion_sort(list1):    
    n = len(list1)
    for i in range(n):
        j = i
        while j > 0 and list1[j-1] > list1[j]:
            temp = list1[j]
            list1[j] = list1[j-1]
            list1[j-1] = temp
            j = j-1
    return list1

# Ask user for input
input_string = input("Please enter your string: ")

#Sort the string in alphabetical order using the simple_sort function
sorted_list = insertion_sort(list(input_string))

# Convert from a list to a string
sorted_string = ""
for i in sorted_list:
    sorted_string += i

#Print the answer
print(sorted_string)
```
