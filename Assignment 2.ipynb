```python
# i) & iv)

# Write a function that asks the user for 5 unique strings one at a time using a For loop and stores them in a set. The function must return a set with 5 values

def unique_strings():
    
    """
    This function prompts the user to enter 5 unique strings

    The are no required arguments for this function

    The return value is a set containing 5 unique strings entered by the user
    """

    # Define empty set to store unique strings
    unique_strings_set = set()
    
    # For loop to ask user for input 5 times
    for i in range(5):
        user_input = input("Please enter a unique string: ")
       
       # Check if input is already in the set
        while user_input in unique_strings_set:
            user_input = input("That string has already been entered. Please enter a unique string: ")
        
        # Add the user input to the set
        unique_strings_set.add(user_input)
    
    # Return the set of unique strings
    return unique_strings_set

# Call the function
unique_strings()
```

```python
# ii) & iv)

# Write a function that accepts a set 'words' and an integer 'cutoff' and returns a set containing the elements of 'words' that are shortened to be no longer than the 'cutoff'.

def shorten_words(words, cutoff):
    
    """
    This function takes a set of words and shortens each word where relevant to be no longer than a specified cutoff length

    The two arguments required for this function is; a set of words to potentially shorten, and an integer to define the cutoff length

    The return value is a set containing the elements of the set 'words' that are shortened to be no longer than the number 'cutoff'
    """
    
    
    # Define empty set to store shortened words
    shortened_words = set()
    
    # For loop to iterate through each word in the input set
    for word in words:
        
        # Check if the length of the word is greater than the cutoff
        if len(word) > cutoff:
            
            # If the word is longer than the cutoff, add the shortened version to the new set
            shortened_words.add(word[:cutoff])
        
        else:
            
            # If the word is shorter than or equal to the cutoff, add it to the new set unchanged
            shortened_words.add(word)
    
    # Return the set of shortened words
    return shortened_words

# Call the function 
shorten_words({"hello", "world", "python", "programming", "ace"}, 5)
```

```python
# iii) & iv)

# Write code to run the functions in part i) and ii), with'5' as the 'cutoff' and user's input as 'words'. The result should be printed to the console.

# Call the unique strings function to get a set of unique strings from the user
user_words = unique_strings()

# Call the shorten words function using the user's input from the previous function with a cutoff of 5
shortened_user_words = shorten_words(user_words, 5)

# Print the result of the shortened words to the console
print(shortened_user_words)
```
